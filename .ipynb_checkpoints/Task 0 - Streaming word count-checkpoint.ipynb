## Demonstrate the spark streaming
    . Send
        - From a terminal
    . Receive
        - From spark streaming context
        - A simple word count program

```python
# importing necessary libraries
import sys
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
import os

#setting up the environment
os.environ['SPARK_HOME'] = "C:\Spark_App\spark-3.2.0-bin-hadoop2.7"

if __name__ == "__main__":
    hostname = "localhost"
    port = int(9999)
    sc = SparkContext(appName="PythonStreamingNetworkWordCount")
    ssc = StreamingContext(sc, 10) # setting up time interval for reading streaming input

    lines = ssc.socketTextStream(hostname, port) #reading lines
    counts = lines.flatMap(lambda line: line.split(" "))\
                  .map(lambda word: (word, 1))\
                  .reduceByKey(lambda a, b: a+b)
    counts.pprint()
    ssc.start()
    ssc.awaitTermination()
```
